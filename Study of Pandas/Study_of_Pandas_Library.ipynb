```python
# Create a DataFrame
data = {
    "Name":["Tanmay","Agarwal","Harshit"],
    "Marks":[85,90,78]
}
df = pd.DataFrame(data)
df
```

```python
# Display Column Names
df.columns
```

```python
# Check Data type of Each Column
df.dtypes
```

```python
# Access Individual Column
print(df["Name"])
df["Marks"]
```

```python
# Access specific row
df.loc[1]
```

```python
# Add a new column
'''If marks 80-90-First class
If marks>=90-Distinction
If Marks <80-Second Class'''
df["Grade"] = ["First Class","Distinction","Second Class"]
df
```

```python
# Update the data in the data frame
# Change marks of student A to 88.
df.loc[0,"Marks"]=98
df.loc[0,"Grade"]="Distinction"
df
```

```python
# Delete a Column
df.drop("Grade",axis = 1, inplace = True)
df
```

```python
# Perform Basic Statistical Analysis
mean1=df["Marks"].mean()
mean1
```

```python
# Mode (Not applicable when you have integer value)
mode1=df["Marks"].mode()
print(mode1)
```

```python
#Find minimum and maximum value of given data set
min1=df["Marks"].min()
print(min1)
max1=df["Marks"].max()
print(max1)
```

```python
# Apply condition (Filtering)
# Display students scoring more than 80
df[df["Marks"]>80]
```
