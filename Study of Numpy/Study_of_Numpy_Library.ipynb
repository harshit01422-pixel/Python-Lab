```python
# Declaration of numpy
import numpy as np
a = np.array([10,20,30,40])
b = np.array([[1,2,3],[4,5,6]])
print(a)
print(b)
```

```python
# Dimension of array
print(a.ndim)
print(b.ndim)
```

```python
# Find out shape of array
print(a.shape)
print(b.shape)
```

```python
# Find out data type of array
print(a.dtype)
print(b.dtype)
```

```python
# In Built functions using numpy
np.zeros((3,2))
```

```python
# Creates symmetric matrix
np.ones((3,3))
```

```python
# Creates diagonal matrix
np.eye((5))
```

```python
np.arange(1,10,2)
```

```python
np.linspace(0,1,4)
```

```python
print(b*2)
print(a+5)
print(a)
print(b)
```

```python
# Mean of array
print(np.mean(a))
print(np.mean(b))
```

```python
# Median of array (when median and mean are same, you have no outliers and data is clean. But when their difference is large, the data is not clean )
print(np.median(a))
print(np.median(b))
```

```python
# Find maximum
print(np.max(a))
print(np.max(b))
```

```python
# Find minimum
print(np.min(a))
print(np.min(b))
```

```python
# Find sum
print(np.sum(a))
print(np.sum(b))
```
