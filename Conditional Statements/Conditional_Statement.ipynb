```python
# Check for Leap Year
year = int(input("Enter Year :"))
if year%4==0 and year%100!=0 or year%400==0:
  print("Leap Year")
else :
  print("Not a Leap Year")
```

```python
# Upcoming Leap Year
year = int(input("Enter Current Year :"))
while True:
  if year%4==0 and year%100!=0 or year%400==0:
    print("Next Leap Year",year)
    break
  year+=1
```
